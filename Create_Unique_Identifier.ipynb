```python
import os
import pandas as pd

_FOLDER = "fitted_datasets_CCL/"
_FOLDER_2 = "test_train/"


def process_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):  # Check if the file is a CSV
            file_path = os.path.join(folder_path, filename)
            # Read the dataset
            df = pd.read_csv(file_path)
            
            # Create the DRUGID_COSMICID column
            df['DRUGID_COSMICID'] = df['DRUG_ID'].astype(str) + "_" + df['COSMIC_ID'].astype(str)
            
            # Save the modified DataFrame back to the same file
            df.to_csv(file_path, index=False)
            print(f"Processed and updated {file_path}")

# Process each folder
process_folder(_FOLDER)
process_folder(_FOLDER_2)

print("All datasets have been processed.")
```
