### DrugBank

```python
from collections import defaultdict

def get_properties_drug(drug_id, drug_name):
    
    dict_tables = {}
    not_found_drugs = 0
    
    try:
        q = pd.read_html("https://www.drugbank.ca/unearth/q?utf8=%E2%9C%93&searcher=drugs&query="+drug_name)
    
        
        ingr = 0
        target = 0
        inter = 0 
        prop = 0
        
        for i in range(len(q)):
            if q[i].columns[0] == 'Ingredient':
                ingr+=1
                q[i].to_csv("drugs/"+str(drug_id)+"_"+drug_name+"_Ingredient_"+str(ingr)+".csv")
                dict_tables.setdefault("Ingredient", []).append(i)
    
            elif q[i].columns[0] == 'Target':
                target+=1
                q[i].to_csv("drugs/"+str(drug_id)+"_"+drug_name+"_Target_"+str(target)+".csv")
                dict_tables.setdefault("Target", []).append(i)
        
            elif q[i].columns[0] == 'Interacting Gene/Enzyme':
                inter+=1
                q[i].to_csv("drugs/"+str(drug_id)+"_"+drug_name+"_Interaction_"+str(inter)+".csv")
                dict_tables.setdefault("Interaction", []).append(i)
            
            elif q[i].columns[0] == 'Property':
                prop+=1
                q[i].to_csv("drugs/"+str(drug_id)+"_"+drug_name+"_Property_"+str(prop)+".csv")
                dict_tables.setdefault("Property", []).append(i)
    except:
        print("No found tables for:", drug_id, drug_name)
        not_found_drugs = drug_id
           
    return dict_tables, not_found_drugs
```
